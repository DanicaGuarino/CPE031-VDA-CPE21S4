2. Does smoking  significantly affect the risk of developing heart disease?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with your actual data if available)
data = {'Smoking': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No'],
        'HeartDisease': ['Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No']}
df = pd.DataFrame(data)

# Grouping and counting
counts = df.groupby(['Smoking', 'HeartDisease'])['HeartDisease'].count().unstack()

# Create bar plot
counts.plot(kind='bar', color=['purple', 'green'], figsize=(10, 6))

# Customize plot
plt.title('Relationship between Smoking and Heart Disease')
plt.xlabel('Smoking Status')
plt.ylabel('Number of Individuals')
plt.grid(axis='y')
plt.legend(title='Heart Disease')
plt.xticks(rotation=0)

plt.show()
```

The graph show the relationship between smoking and heart disease. Based on the graph, it shows that there is a high vote for "No" for those who don't smoke while there is a high count of heart diseases for those are smoking. All in all,mthere is a relationship between smoking and Heart Disease.
