#**3. Analyze the data and generate the appropriate plot including its properties:**

a. Read each device's date of release and sales. Show it using a line plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(path, skiprows=1)

df.columns = ["Brand", "Device", "Model", "Country of Origin", "Date of Release", "Sales (USD)"]
df['Date of Release'] = pd.to_datetime(df['Date of Release'], errors='coerce')
df = df.dropna(subset=['Date of Release', 'Sales (USD)'])
df = df.sort_values('Date of Release')

plt.figure(figsize=(12, 6))
plt.plot(df['Date of Release'], df['Sales (USD)'], marker='o', linestyle='-', color='purple')
plt.title('Device Release Dates and Sales')
plt.xlabel('Date of Release')
plt.ylabel('Sales (USD)')
plt.grid()
plt.xticks(rotation=45)
plt.show()
```

b. Get the total sales and show line plot with the following style properties.

    - Line Style dotted and Line-color should be red.
    - Show legendat the lower right location.
    - X label name = Year
    - Y label name = Sales
    - Add a circle marker.
    - Line marker color as red
    - Line width should be 3





```python
df = pd.read_csv(path, skiprows=1)
df.columns = ["Brand", "Device", "Model", "Country of Origin", "Date of Release", "Sales (USD)"]
df['Date of Release'] = pd.to_datetime(df['Date of Release'], errors='coerce')
df = df.dropna(subset=['Date of Release', 'Sales (USD)'])
df['Year'] = df['Date of Release'].dt.year
yearly_sales = df.groupby('Year')['Sales (USD)'].sum()


plt.figure(figsize=(12, 6))
plt.plot(yearly_sales.index, yearly_sales.values,
         linestyle=':', color='red', marker='o', markerfacecolor='red',
         linewidth=3, label='Total Sales')


plt.title('Yearly Total Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

c. Read all product sales data and show it using a multiline plot. Label them.

```python
df.columns = ["Brand", "Device", "Model", "Country of Origin", "Date of Release", "Sales (USD)"]
df['Date of Release'] = pd.to_datetime(df['Date of Release'], errors='coerce')
df = df.dropna(subset=['Date of Release', 'Sales (USD)'])
df = df.sort_values('Date of Release')

plt.figure(figsize=(12, 8))

for device, group in df.groupby('Device'):
    plt.plot(group['Date of Release'], group['Sales (USD)'], marker='o', linestyle='-', label=device)

plt.title('Sales Data for Each Product')
plt.xlabel('Date of Release')
plt.ylabel('Sales (USD)')
plt.legend(title="Device", loc='upper left', bbox_to_anchor=(1.05, 1))
plt.grid()
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

d. Read each brand sales data and show it using the bar chart. Label them


```python
df.columns = ["Brand", "Device", "Model", "Country of Origin", "Date of Release", "Sales (USD)"]
df['Sales (USD)'] = pd.to_numeric(df['Sales (USD)'], errors='coerce')
df = df.dropna(subset=['Sales (USD)', 'Brand'])
brand_sales = df.groupby('Brand')['Sales (USD)'].sum()


plt.figure(figsize=(10, 6))
brand_sales.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Total Sales by Brand')
plt.xlabel('Brand')
plt.ylabel('Total Sales (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

e. Read each device and show it using the histogram to see the most common Country of Origin. Label them.


```python
df.columns = ["Brand", "Device", "Model", "Country of Origin", "Date of Release", "Sales (USD)"]
df = df.dropna(subset=['Country of Origin'])

plt.figure(figsize=(10, 6))
plt.hist(df['Country of Origin'], bins=len(df['Country of Origin'].unique()), color='magenta', edgecolor='black')


plt.title('Frequency of Devices by Country of Origin')
plt.xlabel('Country of Origin')
plt.ylabel('Number of Devices')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

f. Create a pie chart that shows each brand Country of Origin. Label them.

```python

df.columns = ["Brand", "Device", "Model", "Country of Origin", "Date of Release", "Sales (USD)"]
df = df.dropna(subset=['Country of Origin'])

country_counts = df['Country of Origin'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Proportion of Devices by Country of Origin')
plt.show()
```
