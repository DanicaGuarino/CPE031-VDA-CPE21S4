1. Is there a relationship between Alcohol Consumption and Heart Attack History?

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import pandas as pd

cross_tab = pd.crosstab(df['HadHeartAttack'], df['AlcoholDrinkers'])
colors = ['green', 'purple']

cross_tab.plot(kind='bar', stacked=True, figsize=(8, 6), color=colors)
plt.title('Relationship between Alcohol Consumption and Heart Attack History')
plt.xlabel('Had Heart Attack')
plt.ylabel('Count')
plt.legend(title='Alcohol Consumption')
plt.show()
```

**Observation:** Based on the bar chart visualizing the relationship between alcohol consumption and heart attack history, those who use alcohol tend to have a higher risk of having a heart attack than those who do not consume alcohol. The graph indicates a greater number of people with a history of heart attacks in the "Yes" category for consuming alcohol.


2. How does BMI vary across different sleep categories?

```python
mean_bmi_by_sleep_hours = df.groupby('SleepHours')['BMI'].mean()

plt.figure(figsize=(10, 6))
plt.bar(mean_bmi_by_sleep_hours.index, mean_bmi_by_sleep_hours.values, color='purple')
plt.title('Mean BMI Variation across Sleep Hours')
plt.xlabel('Sleep Hours')
plt.ylabel('Mean BMI')
plt.show()
```

**Observation:** Based on the bar chart visualizing the mean BMI variation across different sleep hour categories, shows a minor trend indicating that those who sleep fewer or more hours may have a higher average BMI than those with average sleep lengths. Individuals reporting an average sleep duration of 6-8 hours tend to have a lower BMI.
