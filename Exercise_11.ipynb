```python
import numpy as np
import matplotlib.pyplot as plt


# Given data points
x = np.array([0.50, 1.00, 1.30, 1.60, 2.20, 2.50, 3.10, 3.90, 4.40]).reshape(-1, 1)
y = np.array([1.90, 1.50, 1.20, 1.00, 0.78, 0.65, 0.46, 0.30, 0.23])

# Perform linear regression
slope, intercept = np.polyfit(x.flatten(), y, 1)

# Calculate predicted values
y_pred = slope * x.flatten() + intercept

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, y_pred, color='red', label='Linear Model')
plt.title("Linear Regression Model")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
```

**INTERPRETATION**

The graph shows that as x increases, y decreases at a consistent rate. It implies that there is a steady reduction in y as x grows. It shows negative correlation.
