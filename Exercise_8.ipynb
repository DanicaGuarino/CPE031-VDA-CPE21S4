Matplot Labels and Title - You can use the xlabel() and ylabel()

```python
import matplotlib.pyplot as plt
import numpy as np

#plotting x and y points
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle= '-.')
plt.show()
```

```python
xpoints=np.array([1,2,6,8]) #plotting x and y points
ypoints=np.array([3,8,1,10])

plt.plot(xpoints,ypoints) #plot() is used to draw points

plt.title ("Distance Travelled Everyday", color="magenta")
plt.xlabel ("Distance in km/s") #labels
plt.show()
```

```python
xpoints=np.array([1,2,6,8]) #plotting x and y points
ypoints=np.array([3,8,1,10])

plt.plot(xpoints,ypoints) #plot() is used to draw points

plt.title ("Distance Travelled Everyday")
plt.xlabel ("Distance in km/s") #labels
plt.ylabel("Speed in km/s)")
plt.grid()
plt.show()
```

```python
xpoints=np.array([1,2,6,8]) #plotting  x and y points
ypoints=np.array([3,8,1,10])

plt.plot(xpoints,ypoints) #plot() is used to draw points
plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km") #labels
plt.ylabel("Speed in km/s")
plt.grid(axis='x', linestyle='--',linewidth=0.50)
plt.show()
```

Show Multiple Plots

```python
#plot 1
xpoints=np.array([1,2,6,8]) #plotting  x and y points
ypoints=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(xpoints,ypoints) #plot() is used to draw points

#plot 2
xpoints=np.array([1,2,6,8]) #plotting  x and y points
ypoints=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(xpoints,ypoints) #plot() is used to draw points

plt.grid(axis='x',linestyle='--',linewidth=0.50)
plt.show()
```

Matplot Scatter- The scatter function plots one dot for each observation

```python
xpoints=np.array([1,2,6,8]) #plotting  x and y points
ypoints=np.array([3,8,1,10])

plt.scatter(xpoints,ypoints)

xpoints=np.array([1,2,6,8]) #plotting  x and y points
ypoints=np.array([10,20,30,40])

plt.scatter(xpoints,ypoints)

plt.show()
```

Matplotlib Bars


*  You can use the bar() to draw bars


```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array(['Student1', 'Student 2', 'Student 3']) #plotting  x and y points
y=np.array([90, 85, 86])

plt.bar(x,y)
plt.show()
```

Matplotlib Pie Chart - You can use pie() to draw pie chart

```python
g = np.array([35, 25, 25, 15])
grader=["Grade 1=35%", "Grade 2=25%", "Grade 3=25%", "Grade 4=15%"]
plt.pie(g,labels=grader)
plt.legend(title ="Grader", labels=grader)
plt.show()
```
